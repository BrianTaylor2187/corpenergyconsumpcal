```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.geocoders import Nominatim
import keys
```

```python
df = pd.read_csv('Corporate_Energy_Consumption.csv', low_memory=False)
```

```python
df.head()
```

```python
latlongset = df['FacilityAddress'].unique()
```

```python
latlongset = pd.DataFrame(latlongset, columns=['FacilityAddress'])
latlongset.head()
```

```python
from geopy.geocoders import ArcGIS, Bing, Nominatim, OpenCage, GoogleV3, OpenMapQuest
```

```python
arcgis = ArcGIS(timeout=100)
nominatim = Nominatim(user_agent="CalLatLongList", timeout=100)
```

```python
geocoders = [arcgis, nominatim]
```

```python
def gc(address):
    street = str(address['FacilityAddress'])
    city = str('Calgary')
    add_concat = street + ", " + city
    for gcoder in geocoders:
        location = gcoder.geocode(add_concat)
        if location != None:
            print(f'geocoded record {address.name}: {street}')
            located = pd.Series({
                'lat': location.latitude,
                'lng': location.longitude,
                'time': pd.to_datetime('now')
            })
        else:
            print(f'failed to geolocate record {address.name}: {street}')
            located = pd.Series({
                'lat': 'null',
                'lng': 'null',
                'time': pd.to_datetime('now')
            })
        return(located)
```

```python
latlongset = latlongset.merge(latlongset.apply(lambda add: gc(add), axis=1), left_index=True, right_index=True)
```

```python
latlongset[latlongset['FacilityAddress'].isnull() == True]
```

```python
df = pd.merge(df, latlongset, on='FacilityAddress', how='outer')
```

```python
df.drop('time', axis=1,inplace=True)
```
